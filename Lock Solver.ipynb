```python
def func_1(num,pattern):
    #Enter the one number is correct and well placed
    patterns = list(pattern)
    value = list(num)
    # either a or b or c contains the value of actual_lock
    for i in range(3):
        if patterns[i] == value[i]:
            correct_value = patterns[i]
            incorrect_value = [x for x in patterns if x not in correct_value]
            check_val = any(item in incorrect_value for item in value)
            if check_val == False:
                return True
                break
    

def func_2(num,pattern):

    patterns = list(pattern)
    value = list(num)
    for i in range(3):
        for j in range(3):
            if patterns[i] == value[j] and i != j:
                correct_value = patterns[i]
                incorrect_value = [x for x in patterns if x not in correct_value]
                check_val = any(item in incorrect_value for item in value)
                if check_val == False:
                    return True
                    break
                
    # one no is correct but wrongly placed
    
def func_3(num,pattern):
    
    patterns = list(pattern)
    value = list(num)
    count = 0
    for i in range(3):
        for j in range(3):
            if patterns[i] == value[j] and i != j:
                count += 1
                if count == 2:
                    return True
    # two no is correct and wrongly placed

def func_4(num,pattern):
    patterns = list(pattern)
    return operator.not_(any(item in patterns for item in list(num)))
    # nothing is correct
    
def func_5(num,pattern):
    patterns = list(pattern)
    value = list(num)
    for i in range(3):
        for j in range(3):
            if patterns[i] == value[j] and i != j:
                correct_value = patterns[i]
                incorrect_value = [x for x in patterns if x not in correct_value]
                check_val = any(item in incorrect_value for item in value)
                if check_val == False:
                    return True
                    break
    
    
    # one no is correct but wrongly placed


    
        
        
    
```
